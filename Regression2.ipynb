```python
import pandas as pd
df=pd.read_csv("/content/dataset.csv")
df.head()
```

```python
df.drop(['Unnamed: 0','Name'],axis=1,inplace=True)
```

```python
df['Location'].value_counts()
```

```python
df['Fuel_Type'].value_counts()
```

```python
df['Owner_Type'].value_counts()
```

```python
df['Transmission'].value_counts()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
dumm=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']])
```

```python
dumm
```

```python
dumm.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in dumm.columns:
  dumm[i]=le.fit_transform(dumm[i])
```

```python
dumm.head()
```

```python
df.drop(['Location','Fuel_Type','Transmission','Owner_Type'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df1=pd.concat([df,dumm],axis=1) #combine df and dumm (axis=1 combining vertically )
```

```python
df1.head()
```

```python
df.isnull().sum()
```

```python
df1['Mileage']=df1['Mileage'].str.replace("km/kg"," ") #remove units in mileage column
```

```python
df1.head()
```

```python
df1['Mileage']=df1['Mileage'].str.replace("kmpl"," ")
```

```python
df1['Engine']=df1['Engine'].str.replace("CC"," ")
```

```python
df1.head()
```

```python
df1['Power']=df1['Power'].str.replace("bhp"," ")
```

```python
df1.head()
```

```python
df1['Power']=df1['Power'].str.replace("null","0")
```

```python
df1.head()
```

```python
df1['Mileage']=df1['Mileage'].astype(float) #change the datatype string to float
```

```python
df1['Engine']=df1['Engine'].astype(float)
```

```python
df1['Power']=df1['Power'].astype(float)
```

```python
df1.info()
```

```python
df1.isna().sum()
```

```python
df1['Mileage'].fillna(df1['Mileage'].mean(),inplace=True)
```

```python
df1.isna().sum()
```

```python
df1['Engine'].fillna(df1['Engine'].mean(),inplace=True)
```

```python
df1['Power'].fillna(df1['Power'].mean(),inplace=True)
```

```python
df1['Seats'].fillna(df1['Seats'].mode()[0],inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1.head()
```

```python
y=df1['Price']
y
```

```python
x=df1.drop(['Price'],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predict=lr.predict(x_test)
```

```python
y_predict
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error,mean_squared_error
print(r2_score(y_test,y_predict))
print(mean_absolute_error(y_test,y_predict))
print(root_mean_squared_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
```
