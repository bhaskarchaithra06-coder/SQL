```python
import pandas as pd
df=pd.read_csv("/content/Mall_Customers (1).csv")
df.head()
```

```python
df.isna().sum()
```

```python
df=df.iloc[:,3:]
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(df)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
km=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_predict=km.fit_predict(df)
```

```python
y_predict
```

```python
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],c=y_predict)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='red')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df,y_predict)
```
