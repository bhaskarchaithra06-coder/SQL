```python
flat_data=[]
target=[]
categories=["cats_set","dogs_set"]
file_path="/content/drive/MyDrive/archive (14)"
for i in categories:
  print("loading.....",i)
  path=os.path.join(file_path,i) #cat set path

  for img in os.listdir(path): #iterating each image in cat set
    img_array=imread(os.path.join(path,img))
    img_resize=resize(img_array,(150,150,3))
    flat_data.append(img_resize.flatten())
    target.append(categories.index(i))
    print("loaded")
```

```python
df=pd.DataFrame(flat_data)
df.head()
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flat_data,target,test_size=.2,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
sv1=SVC(kernel="linear")
sv2=SVC(kernel="poly")
sv3=SVC(kernel="rbf")
l=[sv1,sv2,sv3]
for i in l:
  i.fit(x_train,y_train)
  y_predict=i.predict(x_test)
  print(accuracy_score(y_test,y_predict))
```

```python
import pandas as pd
df=pd.read_csv("/content/Iris.csv")
df.head()
```

```python
df.drop(["Id"],axis=1,inplace=True)
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
from sklearn.svm import SVC
sv=SVC(probability=True,kernel="linear")
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="entropy",max_depth=3)
```

```python
from sklearn.ensemble import VotingClassifier
vc=VotingClassifier(estimators=[("knn",knn),("sv",sv),("nb",nb),("dt",dt)],voting="soft")
vc.fit(x_train,y_train)
y_predict=vc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
```

```python
import pandas as pd
df=pd.read_csv("/content/glass.csv")
df.head()
```

```python
import seaborn as sns
sns.countplot(x=df["Type"])
```

```python
df.corr()
```

```python
df.drop(["K","Ca"],axis=1,inplace=True)
```

```python
sns.heatmap(df.corr())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier
x=df.iloc[:,:7]
y=df.iloc[:,-1]
sc=StandardScaler()
sc.fit(x)
x=sc.transform(x)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

adb_clf=AdaBoostClassifier(n_estimators=10)
adb_clf.fit(x_train,y_train)
y_pred = adb_clf.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred)*100)
```
